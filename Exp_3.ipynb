```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Sample data
data={'category':['A','B','C','D'],'value':[10,20,15,25]}
df=pd.DataFrame(data)
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='category',y='value',data=df)
plt.title('Vertical Bar Plot')
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.barplot(y='category', x='value', data=df)
plt.title('Horizontal Bar Plot')
plt.xlabel('Value')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
data = {
    'category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'group': ['G1', 'G2', 'G1', 'G2', 'G1', 'G2'],
    'value': [10, 15, 20, 18, 12, 14]
}
df = pd.DataFrame(data)

# Grouped bar plot
plt.figure(figsize=(8,6))
sns.barplot(x='category', y='value', hue='group', data=df)
plt.title('Grouped Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.legend(title='Group')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'group': ['G1', 'G2', 'G1', 'G2', 'G1', 'G2'],
    'value': [10, 15, 20, 18, 12, 14]
}
df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot(index='category', columns='group', values='value')
```

```python
pivot_df.plot(kind='bar', stacked=True, figsize=(8,6), color=['skyblue', 'salmon'], edgecolor='black')

plt.title('Stacked Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.legend(title='Group')
plt.tight_layout()
plt.show()
```
