```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/house_prices.csv")
print(df.head)
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(12,6))
avg_price_by_condition = df.groupby('condition')['price'].mean()

plt.bar(avg_price_by_condition.index, avg_price_by_condition.values, color='skyblue', edgecolor='black')
plt.title('Average House Price by Condition')
plt.xlabel('Condition')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.hist(df['price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
