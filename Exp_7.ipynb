```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
```

```python
x=np.random.rand(100)
y=2*x+np.random.normal(scale=0.1,size=100)

data=pd.DataFrame({'Feature1':x,'Feature2':y})
data
```

```python
sns.scatterplot(data=data,x='Feature1',y='Feature2')
plt.title('Scatter Plot with linera relationship')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```
