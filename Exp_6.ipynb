```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/content/Online Retail.csv")
```

```python
df.head()
```

```python

print(df.columns)
```

```python
df['Data']=pd.to_datetime(df['CustomerID'])
df.time_series=df.groupby('CustomerID')['Totalsale'].sum()
```

```python


plt.figure(figsize=(12, 6))
plt.plot(df.time_series.index, df.time_series.values, marker='o', linestyle='-', color='blue')
plt.title("Total Amount Over Time")
plt.xlabel("Date")
plt.ylabel("Total Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```
